6. Explain the concept of generators in Python and how they are defined.
Ans:-
*   A generator in Python is a special type of iterator that allows us to iterate over a sequence of values lazily, meaning it produces items one at a time only when needed, instead of storing the entire sequence in memory at once.
*   Generators are defined in two ways:
1.   Using a function with the yield keyword
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
counter = count_up_to(3)
print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
->print(next(counter))  # Raises StopIteration

->Calling count_up_to(3) does not run the function immediately.
->It returns a generator object.
->Each call to next() executes the function up to the next yield.

2.   Using generator expressions (similar to list comprehensions)
*   This is a compact syntax similar to list comprehension, but uses parentheses instead of brackets:
*   squares = (x * x for x in range(5))

    print(next(squares))  # 0
    print(next(squares))  # 1

7. What are the advantages of using generators over regular functions?

Ans:-
1. Memory Efficiency
->Generators do not store the entire result in memory.

->They yield one item at a time, which is ideal for large datasets or infinite sequences.

-> Example:Instead of returning a list of 1 million numbers, a generator yields one at a time, using constant memory.

 2. Lazy Evaluation
->Generators produce values only when needed.

->This is called lazy evaluation, and it speeds up programs by avoiding unnecessary computations.

-> Example:
If only  first 5 results out of a million are needed, the generator won't compute all million—just the 5.

3. Improved Performance
->Since generators yield items one by one, they can start producing output immediately without waiting to compute everything.

-> Result: Faster response time in programs that don’t need all data at once.

 4. State Retention
->Generators automatically retain their state between yields.

-> don’t need to manually manage loop counters or temporary variables.

 5. Simplified Code
->Generator functions often require less boilerplate code than building a class with __iter__() and __next__().

-> Cleaner alternative to writing a full iterator class.

6. Infinite Sequences
->Generators are ideal for creating infinite streams of data, such as counters, sensor readings, or live data feeds.
def infinite_counter():
    n = 1
    while True:
        yield n
        n += 1
7. Can Replace Complex Iterators
Generator functions can often replace custom iterator classes with just a few lines of code.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:-These three functions—map(), filter(), and reduce()—are functional programming tools in Python. They process iterables (like lists or tuples) and return transformed results, but they do it in different ways.

1. map()
*   Purpose: Applies a function to each item in an iterable and returns a new iterable (map object).
*   Use case: Transforming data.
*   Example:nums = [1, 2, 3, 4]
            squares = list(map(lambda x: x * x, nums)) # Output: [1, 4, 9, 16]
 2. filter()
*   Purpose: Filters elements based on a condition (boolean function).
*   Use case: Extracting data that meets a condition.
*   Example:nums = [1, 2, 3, 4, 5]
            evens = list(filter(lambda x: x % 2 == 0, nums))#Output[2,4]
3. reduce() (from functools module)
*   Purpose: Applies a function cumulatively to the elements of an iterable, reducing it to a single value.
*   Use case: Aggregating data (e.g., sum, product, max).
*   Example:from functools import reduce
            nums = [1, 2, 3, 4]
            product = reduce(lambda x, y: x * y, nums)# Output: 24 (i.e., 1*2*3*4)

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n):
        self.n = n        # total numbers to generate
        self.count = 0    # how many numbers generated so far
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

value=int(input("Enter the number of terms: "))
fib = FibonacciIterator(value)
for num in fib:
    print(num, end=' ')
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_gen(n):
  for i in range(n+1):
    yield 2**i

number=int(input("Enter the number : "))
for num in power_gen(number):
  print(num)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
from google.colab import files
def create_file():
  with open("example.txt", "w") as file:
    file.write("Hello, this is a text file created in Google Colab!\n")
    file.write("You can write multiple lines like this.\n")

def read_file():
  with open("example.txt", "r") as file:
    content = file.read()
    yield content

create_file()
for content in read_file():
  print(content)
```
